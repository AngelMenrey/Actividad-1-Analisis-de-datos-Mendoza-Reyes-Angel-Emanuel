1. Teléfono con mayor capacidad de batería y más económico

```python
import pandas as pd

df = pd.read_csv("mobiles.csv", encoding='latin1')

df["Battery Capacity"] = df["Battery Capacity"].str.replace('mAh', '').str.replace(',', '').astype(float)
df["Launched Price (USA)"] = df["Launched Price (USA)"].str.replace('USD ', '').str.replace(',', '').astype(float)

max_battery_capacity = df["Battery Capacity"].max()
phones_with_max_battery = df[df["Battery Capacity"] == max_battery_capacity]

phones_with_max_battery_sorted = phones_with_max_battery.sort_values(by="Launched Price (USA)")

from IPython.display import display, HTML

print("Teléfonos con mayor capacidad de batería ordenados por precio:")
display(HTML(phones_with_max_battery_sorted.to_html(index=False)))
```

2. Teléfonos con pantalla más pequeña y mayor precio

```python
import pandas as pd

df = pd.read_csv("mobiles.csv", encoding='latin1')

df["Screen Size"] = pd.to_numeric(df["Screen Size"].str.replace(' inches', ''), errors='coerce')
df["Launched Price (USA)"] = pd.to_numeric(df["Launched Price (USA)"].str.replace('USD ', '').str.replace(',', ''), errors='coerce')

df.dropna(subset=["Screen Size", "Launched Price (USA)"], inplace=True)

min_screen_size = df["Screen Size"].min()
phones_with_min_screen = df[df["Screen Size"] == min_screen_size]

phones_with_min_screen_sorted = phones_with_min_screen.sort_values(by="Launched Price (USA)", ascending=False)

from IPython.display import display, HTML

print("Teléfonos con pantalla más pequeña y mayor precio:")
display(HTML(phones_with_min_screen_sorted.to_html(index=False)))
```

5. Teléfonos con mayor precio y más pesados


```python
import pandas as pd

df = pd.read_csv("mobiles.csv", encoding='latin1')

df["Launched Price (USA)"] = pd.to_numeric(df["Launched Price (USA)"].str.replace('USD ', '').str.replace(',', ''), errors='coerce')
df["Mobile Weight"] = pd.to_numeric(df["Mobile Weight"].str.replace('g', ''), errors='coerce')

df.dropna(subset=["Launched Price (USA)", "Mobile Weight"], inplace=True)

phones_sorted_by_price = df.sort_values(by="Launched Price (USA)", ascending=False)

phones_sorted_by_price_and_weight = phones_sorted_by_price.sort_values(by=["Launched Price (USA)", "Mobile Weight"], ascending=[False, False])

from IPython.display import display, HTML

print("Teléfonos con mayor precio y más pesados:")
display(HTML(phones_sorted_by_price_and_weight.to_html(index=False)))
```
